```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('/content/insurance.csv')
print(data)
```

```python
ages = data['age']
smoking_status = data['smoker'].astype('category')
list_example = [ages, smoking_status]
matrix_example = data[['age', 'bmi', 'expenses']].to_numpy()
data_frame_example = data[['age', 'smoker', 'expenses']]
print(ages)
print(smoking_status)
print(list_example)
print(matrix_example)
print(data_frame_example)
```

```python
plt.scatter(data['age'], data['smoker'].apply(lambda x: 1 if x == 'yes' else 0), c='blue', marker='o')
plt.title('Scatter Plot: Age vs Smoking')
plt.xlabel('Age')
plt.ylabel('Smoking (1 = Yes, 0 = No)')
plt.grid(True)
plt.show()
```

```python
filtered_data = data[(data['age'] >= 20) & (data['age'] <= 40)]
plt.scatter(filtered_data['age'], filtered_data['smoker'].apply(lambda x: 1 if x == 'yes' else 0), c='green', marker='+')
plt.title('Scatter Plot: Age (20-40) vs Smoking')
plt.xlabel('Age')
plt.ylabel('Smoking (1 = Yes, 0 = No)')
plt.grid(True)
plt.show()
```

```python
smokers_expenses = data.groupby('smoker')['expenses'].mean()
smokers_expenses.plot(kind='bar', color=['red', 'blue'])
plt.title('Average Expenses: Smokers vs Non-Smokers')
plt.xlabel('Smoker')
plt.ylabel('Average Expenses')
plt.grid(axis='y')
plt.show()
```

```python
smokers_gender = data[data['smoker'] == 'yes'].groupby('sex').size()
smokers_gender.plot(kind='bar', color=['purple', 'orange'])
plt.title('Number of Smokers by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Smokers')
plt.grid(axis='y')
plt.show()
```

```python
plt.scatter(data['age'], data['expenses'], c='orange', marker='x')
plt.title('Scatter Plot: Age vs Expenses')
plt.xlabel('Age')
plt.ylabel('Expenses')
plt.grid(True)
plt.show()
```

```python
teenage_smokers = data[(data['age'] >= 13) & (data['age'] <= 19) & (data['smoker'] == 'yes')]
num_teenage_smokers = teenage_smokers.shape[0]
print(f"Number of teenage smokers: {num_teenage_smokers}")
```
